### Задача 2
Проверьте есть ли у сайта SSL сертификат с использованием модуля requests

```python
import requests

def verification_ssl(url):
    try:
        connection = requests.get(url)
        if connection.url.startswith("https://"):
            print("SSL-сертификат есть")
        else:
            print("SSL-сертификата нет")
    except (requests.exceptions.MissingSchema, requests.exceptions.InvalidURL):
        print("Неверный формат URL")
    except requests.exceptions.SSLError:
        print("SSL-сертификата нет")


verification_ssl("https://yandex.ru")
```

### Задача 3
Выведете при помощи модуля requests блоки 
Status Code, Headers, Url, History, Encoding, Reason, Cookies, Elapsed, Request и Content с сайта https://python.org.

```python
import requests

url = "https://yandex.ru"
connection = requests.get(url)

response_info = {
    "Status Code": connection.status_code,
    "Headers": connection.headers,
    "Url": connection.url,
    "History": connection.history,
    "Encoding": connection.encoding,
    "Reason": connection.reason,
    "Cookies": connection.cookies,
    "Elapsed": connection.elapsed,
    "Request": connection.request,
    "Content": connection.content
}

for k, v in response_info.items():
    print(f"{k}: {v}")
```

### Задача 9
Суммируя весь предыдущий опыт напишите программу, которая будет скрейпить данные из imdb. Рекомендуемые библиотеки: BeautifulSoup, requests и random. Программа должна выполнять следующий функционал:

1. При запуске программа должна подключаться к странице https://www.imdb.com/chart/top
2. Собирать список фильмов
3. Выводить 10 случайных фильмов

```python
import requests
from bs4 import BeautifulSoup
import random

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
url = "https://www.imdb.com/chart/top"

connection = requests.get(url, headers=headers)
soup = BeautifulSoup(connection.text, "html.parser")
titles = [x.text for x in soup.find_all("h3", class_="ipc-title__text")]
titles = titles[1:-1]

for i in range(10):
    random_index = random.randrange(0, len(titles))
    print(titles[random_index])
    del titles[random_index]
```
